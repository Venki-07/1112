```python
df["Sex"]=df["Sex"].replace("female","Female")
df
```

```python
df['Fare']=df['Fare'].astype('int')
df
```

```python
df['Age']=df['Age'].astype('int')
df
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x="Sex",y="Survived",data=df)
plt.title("Sex VS Survived")
plt.xlabel("Gender")
plt.ylabel("Survived")
plt.show()
```

```python
sns.histplot(df['Age'],bins=10,kde=False,hue='Survived')
plt.title("Survivrd based on Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.legend(title='Survived', loc='upper right', labels=['Yes','No'])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='Age', hue='Survived', kde=True)
plt.title('Survival Count based on Socio-Economic Status Age')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['Yes', 'No'])

plt.show()
```

```python
sns.lineplot(x='Pclass' , y='Fare', data=df)
plt.xlabel("Class")
plt.ylabel("Fare")
plt.title("Price VS Class")
plt.show()
```

```python
sns.barplot(x='Survived' , y='Pclass',data=df,color='green',edgecolor='black')
plt.title("Survived based on class in BAR CHART")
plt.xlabel("Survived")
plt.ylabel("Class")
plt.show
```
